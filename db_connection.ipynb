```python
def execute_read_query(connection, query, return_pandas=True):
    cursor = connection.cursor()
    result = None
    try:
        if return_pandas:
            result = pd.read_sql_query(query, connection)
        else:
            cursor.execute(query)
            result = cursor.fetchall()
        return result
    except Error as e:
        print(f"The error '{e}' occurred")
```

```python
select_passengers = """
    SELECT *
    FROM passengers
    LIMIT 10    
"""
passenger_sample = execute_read_query(con, select_passengers)
passenger_sample
```
